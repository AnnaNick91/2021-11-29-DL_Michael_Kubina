```python
# Importiere Bibliotheken

# import urllib (gesamte Bibliothek mit allen Modulen)
# import urllib.request (nur das request-Modul laden)
import urllib.request
# Konvention ist, dass pandas mit pd abgekürzt wird
import pandas as pd
```

```python
# Dataset
url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
file = "openapc.csv"
```

```python
urllib.request.urlretrieve(url, file)
```

```python
# pandas hat viele read-Module... ...wir nutzen read_csv, per <TAB> können wir auch andere "read" durchgucken
apc = pd.read_csv(file)
```
